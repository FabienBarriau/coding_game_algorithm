```python

def toto(exploratory_graph, from_node):
    def loop(actual_node, path_to_root, already_visited):
        while True:
            if exploratory_graph.edges.get(actual_node) is None:
                break
            for new_node, direction in exploratory_graph.edges[actual_node]:
                if new_node in already_visited:
                    continue
                already_visited.add(new_node)
                new_path = path_to_root + [direction]
                if new_node == exploratory_graph.start_node:
                    return new_path
                else:
                    return loop(new_node, new_path, already_visited)
            break

    return loop(from_node, [], {from_node})

toto(exploratory_graph=graph, from_node=(6, 12))
```
